##### Random data ( one dimensional vector) geretated from specific parameter given

```r
(test_sample = rnorm(n = 10, mean = 2, sd = 5))
```

```r
typeof(test_sample)
class(test_sample)
is.vector(test_sample)
```

```r
## Mean std deviation of vector test_sample
mean(test_sample); sd(test_sample)
```

```r
##Standarize it
standardize = function(x) {
  m = mean(x)
  std = sd(x)
  result = (x - m) / std
  result
}
standardize(x = test_sample)
```

```r
### Alternatively above could be wtitten as
standardize = function(x) {
  (x - mean(x)) / sd(x)
```

#### Variance

```r
get_var = function(x, biased = FALSE) {
  n = length(x) - 1 * !biased
  (1 / n) * sum((x - mean(x)) ^ 2)
}
get_var(test_sample)
```

```r
get_var(test_sample)
```

```r
get_var(test_sample, biased = TRUE)# inbuild function for vector
```

```r
var(test_sample)# inbuild function for vector
```
